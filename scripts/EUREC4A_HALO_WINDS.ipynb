```python
###FROM GEET : REMOVE NANS AND FIX GRID (p and z ONLY HALF FREQUENCY COMPARED TO MEASUREMENTS)

#!/usr/bin/env python
# coding: utf-8

# In[ ]:


import xarray as xr
import glob
import numpy as np

directory = foldernc

allfiles = sorted(glob.glob(directory + 'D20200124*_PQC.nc'))
# sorting all files in the directory

ori_list = [None] * len(allfiles)

g = 0

for i in allfiles :
    ori_list[g] = xr.open_dataset(i).dropna(dim='time',subset=['time']).\
    swap_dims({'time':'alt'}).reset_coords().\
    dropna(dim='alt',subset=['alt','pres','u_wind','v_wind','lat','lon','mr'],
                         how='any').\
                  interp(alt=np.arange(0,9100,10))
    g = g + 1


```

```python
#COMBINE ALL SOUNDINGS INTO ONE MATRIX
combined = xr.concat(ori_list[:],dim='sonde_no')
```

```python
#Plot u,v versus altitude


import matplotlib as mpl
from mpl_toolkits import mplot3d
# from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

#Autres timings possibles : 
#11h55
#12h55
#13h55
#14h55
#15h55


# i = 21 # 21 <=> 11h45 outside cold pool
#        # 23<=>11h55 in cold pool
#        # 24 <=> 12h edge cold pool inside cloud
        
for i in [21,23,24]:
        
    ui = combined['u_wind'].isel(sonde_no=i)
    vi = combined['v_wind'].isel(sonde_no=i)
    zi = combined['alt']



    mpl.rcParams['legend.fontsize'] = 10

    # fig = plt.figure()
    # ax = fig.gca(projection='2d')

    # theta = np.linspace(-4 * np.pi, 4 * np.pi, 100) ; z = np.linspace(-2, 2, 100); r = z**2 + 1 ;  x = r * np.sin(theta) ;  y = r * np.cos(theta)

    z = zi/1e3;  x = ui;  y = vi
    # print(z.shape)
    # print(z[1:10]);print((y[1:10]));print((x[1:10]));print((zz[1:10]))
    timei=combined['launch_time'].isel(sonde_no=i)
    timei_label=str(timei.values).split("T")[1].split(".")[0]

    #PLOT ALL HEIGHTS
    plt.figure()
    plt.axhline(y=0,c="k")
    plt.axvline(x=0,c="k")
    plt.scatter(x, y, c=combined['alt'], label='wind')
    plt.colorbar()
    plt.grid(True) 
    plt.title('wind versus height (color in m); time='+timei_label)
    plt.xlabel('u (m/s)')
    plt.ylabel('v (m/s)')
    plt.xlim((-10,25))
    plt.ylim((-14,4))
    plt.savefig("/Users/Caroline/Documents/TALKS/2020-02-Barbados/ColdPools/fig_wind_"+str(i)+".jpg")
    plt.show()
    
    #ZOOM FIRST 2 KMs   
    izmax=200
    plt.figure()
    plt.axhline(y=0,c="k")
    plt.axvline(x=0,c="k")
    plt.scatter(x[0:izmax], y[0:izmax], c=combined['alt'][0:izmax], label='wind')
    plt.colorbar()
    plt.grid(True) 
    plt.title('Zoom: wind versus height (color in m); time='+timei_label)
    plt.xlabel('u (m/s)')
    plt.ylabel('v (m/s)')
    plt.xlim((-10,2))
    plt.ylim((-8,4))
    plt.savefig("/Users/Caroline/Documents/TALKS/2020-02-Barbados/ColdPools/fig_wind_"+str(i)+"_zoom.jpg")
    plt.show()
```
